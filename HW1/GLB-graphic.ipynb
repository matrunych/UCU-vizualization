```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', skiprows=1)
df.head(150)
```

```python
df = df.drop(['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'],  axis=1)
df.head()
```

```python
points_left = df[df['Year'].isin([1904, 1998])]
points_right = df[df['Year'].isin([1944, 2014, 2015, 2016, 2017])]
```

```python
text_points_left = alt.Chart(points_left).mark_text(
        align='left',
        fontSize=15,  
        fontWeight = 200,
        dx=-40
    ).encode(
        x='Year:Q',
        y='J-D:Q',
        text='Year:O'
)

text_points_right = alt.Chart(points_right).mark_text(
        align='right',
        fontSize=15,  
        fontWeight = 200,
        dx=40,
        
    ).encode(
        x='Year:Q',
        y='J-D:Q',
        text='Year:O'
)

text_bold_point = alt.Chart(points_right).mark_text(
        align='right',
        fontSize=15,  
        fontWeight = 'bold',
        dx=40,
    ).encode(
        x='Year:Q',
        y='J-D:Q',
        text='Year:O'
).transform_filter(
    alt.FieldEqualPredicate(field='Year', equal=2016)
)
```

```python
text_upper_line = alt.Chart({'values':[{'Year': 2020, 'Temp': 0.1}]}).mark_text(
        align='right',
        fontSize=15,
    fontWeight = 200,
        text='HOTTER THAN THE \n1880-1899 AVERAGE', 
        lineBreak='\n'
    ).encode(
        x='Year:Q',
        y='Temp:Q'
)

text_under_line = alt.Chart({'values':[{'Year': 2020, 'Temp': -0.05}]}).mark_text(
        align='right',
        fontSize=15,
        fontWeight = 200,
        text='COLDER', 
    ).encode(
        x='Year:Q',
        y='Temp:Q'
)

chart_title = alt.Chart({'values':[{'Year': 1885, 'Temp': 0.99}]}).mark_text(
        align='left',
        fontSize=15,
        fontWeight = 'bold',
    opacity=1,
        text='Annual Global Surface Temperature, \nRelative to Late 19th Century Average',
        lineBreak='\n'
 
    ).encode(
        x='Year:Q',
        y='Temp:Q'
)
```

```python
chart = alt.Chart(df).mark_point(
    filled = True, 
    color = '#F26C08', 
    stroke='#000000', 
    strokeOpacity=0.5,
    size=90).encode(
        x = alt.X(
            field='Year',  
            type = 'quantitative',
            title='',
            scale=alt.Scale(domain=[1880, 2020]),
            axis = alt.Axis(
                 format = "1000f",
                 tickMinStep = 10,
                 tickCount=14,
                 tickColor='white',
                 labelPadding=10,
                 labelFontSize=15,
                 labelOpacity=0.5,
                 labelColor='#999999',
                 domainColor='white',
             ),
        ),
        y = alt.Y(
            field='J-D', 
            type = 'quantitative',
            title='',   
            axis = alt.Axis(
                labelExpr = "format(datum.value, '+.1f') +'°'",
                tickMinStep = 0.2,
                tickCount=9,
                tickColor='grey',
                tickOpacity=0.3,
                labelPadding=10,
                labelFontSize=15,
                labelOpacity=0.5,
                labelColor='#999999',
                domainOpacity=0
            )
        )
)


text = chart.mark_text(
    align='left',
    baseline='middle',
    dx=7
).encode(
    text='Label'
)
```

```python
line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='black', opacity=0.5).encode(y='y')
```

```python
chart_total = alt.layer(chart, line, text_upper_line, text_under_line, chart_title, text_points_left, text_points_right, text_bold_point).configure_view(
    strokeWidth = 0).properties(
    width = 950, height = 600, background = '#FFFFFF', padding = 25)
chart_total
```
